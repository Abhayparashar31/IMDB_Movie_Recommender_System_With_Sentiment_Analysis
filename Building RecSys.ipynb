## Content Based Recommender System

### 1. CountVectorizer (Bag of Words)

```python
# Removing Stop-words
from sklearn.feature_extraction.text import CountVectorizer
countVec = CountVectorizer(max_features=10000)
```

```python
# Tags of Movies Convert in Vectors
Vectors = countVec.fit_transform(df['tags']).toarray()
```

```python
Vectors[0]
```

```python
###### COSING SIMILARITY BETWEEN COUNT VECTORS #########
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(Vectors)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
def get_index_from_title(title):
    return df[df.title == title].index[0]
```

```python
### GENERATING Recommended MOVIES TITLES

user_liked_movie = 'The Avengers'

movie_index = get_index_from_title(user_liked_movie)

similar_movies = list(enumerate(similarity[movie_index]))

sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:11]

print(f'------ Top 10 Recommended Movies For {user_liked_movie}------')
for i in sorted_similar_movies:
    print(get_title_from_index(i[0]))
```

```python
###### SORTING MOVIES BASED ON POPULARITY ##########
user_liked_movie = "Pirates of the Caribbean: At World's End"

movie_index = get_index_from_title(user_liked_movie)

similar_movies = list(enumerate(similarity[movie_index]))

sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:11]

sort_by_average_vote = sorted(sorted_similar_movies,key=lambda x:df["popularity"][x[0]],reverse=True)

print(f'------ Top 10 Recommended Movies For {user_liked_movie}------')
for i in sort_by_average_vote:
    print(get_title_from_index(i[0]))
```

### 2. TF-IDF 

```python
# Removing Stop-words
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfVec = TfidfVectorizer(max_features=10000)
```

```python
# Tags of Movies Convert in Vectors
tfidfVectors = tfidfVec.fit_transform(df['tags']).toarray()
```

```python
tfidfVectors[0]
```

```python
###### COSING SIMILARITY BETWEEN COUNT VECTORS #########
from sklearn.metrics.pairwise import cosine_similarity
similarity_tf = cosine_similarity(tfidfVectors)
```

```python
###### SORTING MOVIES BASED ON popularity #########
user_liked_movie = "Pirates of the Caribbean: At World's End"

movie_index = get_index_from_title(user_liked_movie)

similar_movies = list(enumerate(similarity_tf[movie_index]))

sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:11]

sort_by_average_vote = sorted(sorted_similar_movies,key=lambda x:df["popularity"][x[0]],reverse=True)

print(f'------ Top 10 Recommended Movies For {user_liked_movie}------')
for i in sort_by_average_vote:
    print(get_title_from_index(i[0]))
```

### 3. sigmoid_kernel (Instead of Cosine Similarity)


```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfidfVectors, tfidfVectors)
```

```python
sig[0]
```

```python
###### SORTING MOVIES BASED ON popularity #########

### USER INPUT MOVIE
user_liked_movie = "Pirates of the Caribbean: At World's End"

### MOVIE INDEX
movie_index = get_index_from_title(user_liked_movie)


###  PAIRWISE SIMILARITY SCORES
similar_movies = list(enumerate(sig[movie_index]))

### SORT THE MOVIES AND EXTRACT TOP 10 MOVIES
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:11]

sort_by_average_vote = sorted(sorted_similar_movies,key=lambda x:df["popularity"][x[0]],reverse=True)


print(f'------ Top 10 Recommended Movies For {user_liked_movie}------')
for i in sort_by_average_vote:
    print(get_title_from_index(i[0]))
```
